- Simple Bank Account System:
    - Initial Balance = ₹ 1000.00

#### User-defined Data Structures

```python
class OurBankAccount:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
    
    def withdraw(self, amount):
        if (amount > self.balance):
            print('Insufficient Balance.')
            print(f'{self.name + "'s"} balance: ₹ {self.balance}')
        else:
            self.balance -= amount
            print(f'{self.name} withdrew: ₹ {amount}. New balance: ₹ {self.balance}')
```

```python
user_1 = OurBankAccount('Rubina', 10000)
```

```python
user_1.withdraw(1000)
```

```python
user_1(10000)
```

```python
print(type(user_1))
```

```python
user_2 = OurBankAccount('Deb', 10000)
```

```python
user_2.withdraw(10000)
```

##### Points to remember:
- class Syntax:

        class ClassName:
                def __init__(self, *args, **kwargs):
                        initialize all the necessary variables
                
                def method_2(self, *args, **kwargs):
                        do something
                
                def method_3(self, *args, **kwargs):
                        do something
- Here, the method \_\_init\_\_() initializes automatically whenever a class is called using its object or instance
- Also, \_\_init\_\_() is known as *Constructor*
- \_\_init\_\_() is optional
- \_\_init\_\_() is basically used to initialize or declare variables
- Function inside a *class* is known as *method*. It is also known as *attribute* of the *class*
- We can see all the methods are having the first argument as *self* which creates the link between the methods and the class 
- Every time we create an object it creates a different memory allocation for that particular object

#### Creating A Robot

```python
# Blue-print
class Robot:
    def __init__(self):
        print('Hello, welcome to Robot!')
        self.value_1 = 100
    
    def walk(self):
        print('Walking 10 meters.')
        print(self.value_1)

    def sing(self):
        print('Singing a song.')
        print(self.value_1)
```

```python
obj_1 = Robot()     # returns object
# In terms of memory, obj_1 is assigned with a whole block of space; so the value present at that memory space is known as the object/instance of the class Robot

# an object is the actual or physical data stored in RAM as a package
```

```python
print(type(obj_1))
```

```python
print(id(obj_1))
```

```python
obj_1.walk()
```

```python
obj_1.sing()
```

```python
obj_1.value_1
```

```python
# Blue-print
class Robot:
    def __init__(self, customer_name=None):
        self.name = 'Chitti'
        self.ram_in_gb = 10
        self.version = '0.0.1'
        self.customer_name = customer_name

        print('Hello, welcome to Robot!')
    
    
    def config(self):
        print(f'RAM (in GB) : {self.ram_in_gb}')
        print(f'Version : {self.version}')
        print(f'Robot\'s Name : {self.name}')

    def customer_details(self):
        print(f'Customer Name : {self.customer_name}')
```

```python
r1 = Robot('Rubina')
```

```python
r1.config()
```

```python
r1.customer_details()
```

```python
r2 = Robot('Nikhil')
```

```python
r2.customer_details()
```
