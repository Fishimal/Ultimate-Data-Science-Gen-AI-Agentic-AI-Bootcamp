```python
import time
```

```python
t1 = time.time()        # unit of seconds starting from january 1, 1970 (UNIX time)
```

```python
t1
```

```python
t2 = time.time()
```

```python
t2
```

```python
t2 - t1     # 14 precision points
```

```python
tp_1 = time.perf_counter()      # start is undefined
```

```python
tp_1
```

```python
tp_2 = time.perf_counter()
```

```python
tp_2
```

```python
tp_2 - tp_1     # 15 precision points
```

```python
len('44435906410217')       # time() precision
```

```python
def add(a, b):
    return (a + b)
```

```python
t1_f = time.time()
add(1, 2)
t2_f = time.time()

print(t2_f - t1_f)
```

```python
t1_f_pf = time.perf_counter()
add(1, 2)
t2_f_pf = time.perf_counter()

print(t2_f_pf - t1_f_pf)
```
