### Scope of A Variable

```python
a = 0

def add():
    a += 1
    print(a)
```

```python
add()
```

- Variables defined in a function are called as *local variables* of that function
- Once the function ends, the variable is forgotten

```python
a = 0       # global

def add():
    a = 2       # local
    a += 1
    print(a)


print(a)
add()
print(a)
```

```python
a = [1, 2, 3]

def add():
    a.append(123)
    print(a)


print(a)
add()
print(a)
```

```python
a = [1, 2, 3]

def add():
    a = [4, 5, 6]
    print(a)


print(a)
add()
print(a)
```

```python
a = [1, 2, 3]

def add():
    a = [4, 5, 6]
    a.append(10)
    print(a)


print(a)
add()
print(a)
```

```python
a = [1, 2, 3]

def add():
    a.append(123)
    a = [4, 5, 6]
    a.append(10)
    print(a)


print(a)
add()
print(a)
```

```python
a = 0       # global

# to use a global variable inside a function, we need to declare what global variable we are trying to access and write it inside a function.
def add():
    global a
    a += 1
    print(a)


print(a)        # 0
add()       # 1
print(a)        # 1
```

```python
a = 0       # global

def add():
    print(a)


print(a)
add()
print(a)
```

### Libraries, pip and PyPI

Here is how libraries, pip, and PyPI connect to form the backbone of the Python ecosystem.

Think of it like building a house with Lego blocks: PyPI is the massive online store where people share new Lego pieces, pip is the delivery truck that fetches them for you, and a library is the specific box of Lego blocks you unpack and use in your project.

**1. Libraries**

    A library is a collection of pre-written code, functions, and modules that you can reuse in your own programs so you don't have to reinvent the wheel.

    Instead of writing hundreds of lines of code to handle complex math, web scraping, or data analysis from scratch, you can import a library that already does it perfectly.

    Examples: numpy (for math), requests (for downloading web pages), pandas (for data analysis).


**2. PyPI (Python Package Index)**

    PyPI is the official, free online repository where Python developers from all over the world upload and share their libraries and packages.

    If you write a cool library and want others to use it, you publish it to PyPI. It acts as a massive public catalog of open-source software.


**3. pip**
    
    pip is the standard package manager for Python. It is a command-line tool that connects your computer to PyPI.

    Whenever you want to use a library from PyPI, you use pip to automatically download and install it onto your machine.

```python
def our_join_function(operating_system, *args):
    final_path = ''

    for i in args:
        if (operating_system.lower() == 'windows'):
            final_path += (i + '\\')
        
        if (operating_system.lower() == 'macos'):
            final_path += (i + '/')
    return final_path[:-1]
```

```python
our_join_function('macOS', os.getcwd(), 'temp', 'dummy_2.txt')
```

```python
f = open('dummy.txt')

while True:
    line = f.readline()        # read line by line and moves pointer

    if not line:
        break

    print(line.strip())

f.close()
```
