### Loops
- for loop

#### for loop v/s while loop

| S.No. | for loop | while loop |
| :---: | :--- | :--- |
| 1 | Iterates over a sequence of elements. | Repeats a block of code until certain condition is True. |
| 2 | We already know how many iterations the loop is going to take. | We don't know how many iterations the loop is going to take. |
| 3 | Syntax: for i in [1, 2, 3]: (block of code) | Syntax: while condition_is_True: (block of code) |

#### while loop

```python
# until user writes 'hello', we will keep on taking input from user and print it

user_input = ''

while (user_input != 'hello'):
    user_input = input('Enter your input: ')
    print(user_input)

print('After while loop')
```

```python
# initialize the variable count with value 0
# keep adding +1 to the count until it is smaller than or equal to 5

count = 0
iteration = 1

while (count <= 5):
    print(f'At iteration_{iteration} the value of the counter before increment is {count}')
    count += 1
    print(f'At iteration_{iteration} the value of the counter after increment is {count}')
    iteration += 1

print(f'Total number of iterations = {iteration}')
print(f'Final value of the counter = {count}')
```

### Control Flow Statements or Loop Control Statements

```python
# continue

names = ['sudarshan', 'praveen', 'abhijeet', 'monal', 'gaurav', 'sandeep']
```

```python
for name in names:
    if name == 'monal':
        continue

    name = name.capitalize()
    print(name)
```

```python
for name in names:
    if name == 'monal':
        break

    name = name.capitalize()
    print(name)
```

```python
count = 0

print('START')

while True:     # infinite
    if (count < 100):       # control to exit the loop based on conditions
        break

    print('Hello')
    count += 1

print('END')
```

```python
def add(a, b):
    print(a + b)

def subtract(a, b):
    print(a - b)

def divide(a, b):
    print(a // b)

def multiply(a, b):
    print(a * b)


def calculator():
    choice = int(input('Enter 1 for addition, 2 for subtraction, 3 for divison and 4 for multiplication: '))
    value_1 = float(input('Enter the first value: '))
    value_2 = float(input('Enter the second value: '))

    if (choice == 1):
        print(f'Addition: {value_1} + {value_2}')
        add(value_1, value_2)
    elif (choice == 2):
        print(f'Subtraction: {value_1} - {value_2}')
        subtract(value_1, value_2)
    elif (choice == 3):
        print(f'Floor Division: {value_1} // {value_2}')
        divide(value_1, value_2)
    elif (choice == 4):
        print(f'Multiplication: {value_1} * {value_2}')
        multiply(value_1, value_2)
    else:
        print('Please enter the correct choice to proceed with a specific calculation task')
```
