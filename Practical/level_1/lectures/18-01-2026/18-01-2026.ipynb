### String properties

Whenever we check for the data type of a string in python using the type() function, we can see the output as $\rightarrow$ <class 'str'>

So, from here we need to understand that anything which is a class will always have some properties these properties are known as methods.

Functions are stand-alone code like $\rightarrow$ print(), type(), id(), int(), float(), str(), bool()

Whereas, methods are parts of some other block like methods of 'str' class $\rightarrow$ lstrip(), rstrip(), strip()

```python
print(type(1))
```

```python
print(type('Data'))
```

```python
string = '    Data Science    '
```

```python
print(string.strip())       # strip() is a property or method of class 'str'
```

```python
print(strip(string))        # hence strip() is not a function
```

```python
print(1.strip())        # strip() is not a method of class 'int'
```

```python
print('1'.strip())      # does something but only for its own data type
```

- **method** $\rightarrow$ property of something is called a method
- **function** $\rightarrow$ does some job

```python
a = 2
id(a)       # does something
```

```python
a.id()      # id() is not a property of class 'int'
```

```python
name = 'Rubina D    '
```

```python
# strip(): this method is used to remove all the spaces from left and right side of the string
name = name.strip()
print(name)
```

```python
# lstrip(): this method is used to remove all the spaces from the left side of the string
name = '    Rubina D    '
print(name.lstrip())
```

```python
# rstrip(): this method is used to remove all the spaces from the right side of the string
name = '    Rubina D    '
print(name.rstrip())
```

```python
# remove spaces from both side of the string without using strip()
name = '    Rubina D    '
name = name.lstrip()
name = name.rstrip()
print(name)
```

```python
# remove spaces from both side of the string without using strip()
name = '    Rubina D    '
name = name.lstrip().rstrip()
print(name)
```

```python
# remove spaces from both side of the string without using strip()
name = '    Rubina D    '
name = name.rstrip().lstrip()
print(name)
```
