### Naming Convention

- Variable names can only contain letters, digits or underscore (_)
    - bmi, person0, person_1, _person
- Variable name can't start with a digit
- Variables are case-sensitive : hello != Hello != heLlo
- Avoid using Python keywords : type, print, id, if else, elif, for, while, break

### Operators - I

Operators are used to perform operations on values and variables.

#### Types of Operators:-
1. **Arithmetic Operator** $\rightarrow$ +, -, *, /, //, %, **
2. **Relational Operators** $\rightarrow$ >, <, >=, <=, ==, !=
3. **Assignment Operators** +=, -=, *=, /=, //=, %=, **=
4. **Logical Operators** $\rightarrow$ and, or, not
5. **Bitwise Operators** $\rightarrow$ &, |, ^, ~, <<, >>
