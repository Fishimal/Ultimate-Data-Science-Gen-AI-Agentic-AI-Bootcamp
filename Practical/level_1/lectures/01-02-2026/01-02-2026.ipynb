### File Operations - II

#### Different File Modes

##### Mode Matrix / Table

| Mode | Meaning | Pointer Start | Wipe File | Usage | Creates File if Not Present |
| --- | --- | --- | --- | --- | --- |
| r | Read Only | Beginning | No | Standard Reading | No |
| w | Write Only | Beginning | Yes | Fresh start, deletes old data | Yes |
| a | Append | End | No | Adding more content to the file | Yes |
| r+ | Read + Write | Beginning | No | Modifying existing content | No |
| w+ | Write + Read | Beginning | Yes | Over-writes the file | Yes |
| a+ | Append + Read | End | No | Adding more content to the file and read | Yes |

```python
with open('temp/temp_1.txt', 'r') as f:
    print(f.tell())     # tells the current location of pointer
    all_contents = f.read()
    print(f.tell())     # tells the current location of pointer
    print(all_contents)
```

```python
with open('temp/temp_2.py', 'w+') as f:
    print(f.tell())     # tells the current location of pointer
    f.write("print('hello world')")
    print(f.tell())     # tells the current location of pointer
    f.seek(7)       # moves the cursor to the specified index
    f.write("*****")
    print(f.tell())     # tells the current location of pointer
```

```python
with open('temp/temp_2.txt', 'w+') as f:
    first_index = f.tell()
    print(first_index)     # tells the current location of pointer

    f.write('hello world')

    last_index = f.tell()
    print(last_index)     # tells the current location of pointer

    middle_index = ((last_index // 2) + 1)
    print(middle_index)
    
    f.seek(middle_index)
    f.write("*****")
    print(f.tell())     # tells the current location of pointer
```
