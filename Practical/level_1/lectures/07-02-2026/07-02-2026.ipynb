##### Encapsulation
- Public : {'Syntax' : 'name', 'Access' : 'anywhere'}
- Protected : {'Syntax' : '_name', 'Access' : 'class + child classes'}
- Private : {'Syntax' : '__name', 'Access' : 'only inside class'}

Public, Protected and Private assumes the developer is responsible.

```python
# Scope levels in python

class BankAccount:
    def __init__(self, name, balance):
        self.name = name        # public
        self._account_type = 'Savings'        # protected
        self.__balance = balance        # private
    
    def get_balance(self):
        # getter()
        # to get value of the __balance variable
        print(f'Your current balance : {self.__balance}')
    
    def add_and_set_new_balance(self, amount):
        # setter()
        # to deposit money in our account
        self.__balance += amount
        print(f'Your current balance : {self.__balance}')
```

```python
ba_1 = BankAccount('Shramith', 100000)
```

```python
ba_1.name       # Public Variable -> can be accessed from anywhere
```

```python
ba_1._account_type      # Protected Variable -> warning to tell other developer/s that you should not use the variable outside
```

```python
ba_1.__balance      # Private Variable
```

##### Inheritance
Inheritance allows a class to reuse properties or methods of another class.

**Types of Inheritance**
- *Single-level Inheritance*: $ A(Parent) \xrightarrow{inherits\ from} B(Child) $
- *Multi-level Inheritance*: $ A(Parent) \xrightarrow{inherits\ from} B(Child) \xrightarrow{inherits\ from} B(Grand-child) $
- *Multiple Inheritance*: $ A + B (Parents) \xrightarrow{inherits\ from} C(Child) $

**Syntax:**

class ClassName1:

        def __init__(self, *args, **kwargs):
            do something

        def a_1(self, *args, **kwargs):
            do something

        def a_2(self, *args, **kwargs):
            do something

        def a_3(self, *args, **kwargs):
            do something


class ClassName2(ClassName1):

        def __init__(self, *args, **kwargs):
            do something

        def b_1(self, *args, **kwargs):
            do something

        def b_2(self, *args, **kwargs):
            do something

```python
# Single-level Inheritance

class Device:
    def power_on(self):
        print('Device On')


class TV(Device):
    def show_channels(self):
        print('Channel 1., 2., 3.')
```

```python
tv_1 = TV()
```

```python
tv_1.power_on()
```

```python
tv_1.show_channels()
```

#### Method Overriding
- Same method name

```python
class A:
    def speak(self):
        print('Parent Speak')


class B(A):
    def speak(self):
        # super() -> always refernces to the parent class
        super().speak()     # A().speak()
        print('Child Speak')
```

```python
obj_1 = B()
```

```python
obj_1.speak()
```

```python
class A:
    def __init__(self):
        print('I am in Parent Class A.')
    
    def speak(self):
        print('Parent Speak')


class B(A):
    def __init__(self):
        super().__init__()
        super().speak()
    
    def speak(self):
        print('Child Speak')
```

```python
obj_2 = B()
```

```python
obj_2.speak()
```

Practical Example

```python
class MotherBoard:
    def __init__(self, ram, cpu_speed, cpu_type):
        self.ram = ram
        self.cpu_speed = cpu_speed
        self.cpu_type = cpu_type
    
    def start(self):
        print('Power flow started')
    
    def stop(self):
        print('Power flow stopped')
```

```python
class Computer(MotherBoard):
    def __init__(self, ram, cpu_speed, cpu_type):
        super().__init__(ram, cpu_speed, cpu_type)
        self.os = self.__check_os_and_set_os_type()
    
    def __check_os_and_set_os_type(self):       # private method
        if (self.cpu_type.lower() == 'mac-chip'):
            return 'mac'
        else:
            return 'other'
    
    def check_my_os(self):      # public method
        print(f'User has : {self.os} OS')
```

```python
cmp1 = Computer('12gb', '4.2ghz', 'mac-chip')
```

```python
cmp1.__check_os_and_set_os_type()
```

```python
cmp1._Computer__check_os_and_set_os_type()
```

```python
cmp1.check_my_os()
```

```python
cmp1.start()
```

```python
cmp1.stop()
```

```python
cmp1.ram, cmp1.cpu_speed, cmp1.cpu_type
```

```python
class A:
    def __init__(self):
        print('Class A')
    
    def a(self):
        print('a')


class B(A):
    def __init__(self):
        super().__init__()
        print('Class B')
    
    def b(self):
        print('b')


class C(A):
    def __init__(self):
        super().__init__()
        print('Class C')
    
    def c(self):
        print('c')


class D(B, C):      # not available in Java as it creates ambiguity
    def __init__(self):
        super().__init__()
        print('Class D')
    
    def d(self):
        print('d')
```

```python
d_obj = D()
```

#### Abstraction (Contract)

```python
from abc import ABC, abstractmethod
```

```python
class Payment(ABC):         # requirement no 1

    @abstractmethod         # decorator         # requirement no 2
    def pay(self, amount, currency):        # Arguments matching not mandatory
        pass

    @abstractmethod         # requirement no 2
    def refund(self, amount):        # Arguments matching not mandatory
        pass

# contract called Payment has 2 methods in the contract which all banks need to implement
# first is pay
# second is refund
```

```python
class IPLDBank(Payment):

    def pay(self):
        print('PAY')

    def refund(self):
        print('REFUND')

    def new_user(self):
        print('Hello')
```

```python
obj_b1 = IPLDBank()
```

```python
obj_b1.refund()
```

```python
obj_b1.pay()
```

```python
obj_b1.new_user()
```
