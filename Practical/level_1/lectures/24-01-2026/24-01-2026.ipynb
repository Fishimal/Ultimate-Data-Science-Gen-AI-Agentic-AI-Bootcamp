#### Dictionary (The key-value pair):
1. To store data with labels (keys), instead of positions
2. Key must be (numbers/string/tuples) but values can be anything
3. Keys should be unique, values can be duplicate

##### Ways to create an Empty Dictionary

```python
dictionary_1 = {}
```

```python
dictionary_2 = dict()
```

```python
print(type(dictionary_1), type(dictionary_2))
```

```python
print(dictionary_1, dictionary_2)
```

```python
register = {
    'Rohit': ['12-12-1980', 123456789, 'abcdefgh'],
    'Rohith': ['12-12-1980', 123456789, 'abcd'],
    'Rubina': False
}
```

```python
print(register)
```

```python
government_system = {
    101: ['Rubina'],
    102: ['Vishakha'],
    103: ['Shruti'],
    104: ['Simran']
}
```

##### Accessing elements in a dictionary

```python
government_system[105]
```

```python
government_system.keys()
```

```python
government_system.values()
```

```python
government_system.items()
```

```python
print(list(government_system.keys()))
```

```python
print(list(government_system.values()))
```

```python
print(list(government_system.items()))
```

```python
for key in list(government_system.keys()):
    print(key)
```

```python
for value in list(government_system.values()):
    print(value)
```

```python
for key, value in list(government_system.items()):
    print(f'{key} : {value}')
```

#### Tuple (Immutable):
1. It is used in cases, where data should not be changed (e.g. $\rightarrow$ ID)
2. Prevents accidental cases where data can be overwritten
3. It accepts anything, i.e., values can be of any data type
4. Limitation: append()/pop()/no-reassignment after creation

##### Ways to create an Empty Tuple

```python
tuple_1 = ()
```

```python
tuple_2 = tuple()
```

```python
print(type(tuple_1), type(tuple_2))
```

```python
print(tuple_1, tuple_2)
```

#### Set (The unique values):
1. To handle uniqueness
2. Removes duplicate value automatically
3. It accepts only unique, immutable items
4. It is unordered

```python
guests = set()      # creating an empty set
```

```python
guests
```

```python
print(type(guests))
```

```python
guests = {'Alice', 'Alice', 'Tanmay', 123, 1001, 123, 'Sudarshan'}
```

```python
print(guests)
```

```python
alpha = {'a', 'b', 'd', 'z', 'g'}
```

```python
print(alpha)
```

```python
# how to add elements in a set()

print(guests)
guests.add(10)
print(guests)
```

```python
# use function sorted() to sort elements

sorted(alpha)
```

```python
alpha
```

```python
print(alpha)
```

```python
alpha = sorted(alpha)
```

```python
print(alpha)
```

```python
alpha_reverse = sorted(alpha, reverse=True)
```

```python
print(alpha_reverse)
```

```python
sorted([1, 2, 3, 100, 2])
```
