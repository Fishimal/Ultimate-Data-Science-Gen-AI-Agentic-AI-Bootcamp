**1. The Office Door Badge** You are programming a smart door for a tech office. The door should only open if the user provides the correct "Access Code."
- **Input:** User enters a 4-digit code.
- **Scenario:** If the code is 1234, print "Access Granted. Door Unlocked." Otherwise, print "Access Denied. Security Alerted."

```python
print('The Office Door Badge')

office_door_access_code = 1234
access_code = input('Enter the Access Code to open the door: ').strip()

if (len(access_code) == 4):
    if (int(access_code) == office_door_access_code):
        print('Access Granted. Door Unlocked.')
    else:
        print('Access Denied. Security Alerted.')
else:
    print('Please check the code should be of 4-digits.')
```

**2. The Weekend Grocery Trip** A person is deciding whether to go to the supermarket.

- **Input:** User enters how many eggs are left in the fridge (integer).
- **Scenario:** If the number of eggs is less than 3, print "Time to go to the market!" Else, print "You have

**3. The Smart Home "Leaving for Work" Checklist** When you leave your house, a smart assistant checks multiple
independent things. These are not mutually exclusive.
- **Input:** Three separate inputs (Yes/No) for: "Is the AC on?", "Is the Coffee Maker on?", "Is the Front Door
unlocked?"
- **Scenario:**
    - If AC is on Print "Turning off AC to save energy."
    - If Coffee Maker is on Print "Warning: Coffee Maker is still hot!"
    - If Door is unlocked Print "Locking the front door now." (Note: All three messages could potentially print.)

```python
print('Leaving For Work Checklist')

is_ac_on = input('Is the AC on? (Yes/No): ').strip().capitalize()
is_coffee_maker_on = input('Is the Coffee Maker on? (Yes/No): ').strip().capitalize()
is_front_door_unlocked = input('Is the Front Door Unlocked? (Yes/No): ').strip().capitalize()

if (is_ac_on == 'Yes'):
    print('Turning off AC to save energy.')

if (is_coffee_maker_on == 'Yes'):
    print('Warning: Coffee Maker is still hot!')

if (is_front_door_unlocked == 'Yes'):
    print('Locking the front door now.')

print('No more action required.')
```

**4. The Pizza Topping Calculator** A customer is choosing toppings. Each topping adds a specific cost regardless of
the other toppings chosen.
- **Input:** Three boolean (True/False) inputs for: extra_cheese , pepperoni , and olives.
- **Scenario:** Initialize a total_bill = 10.
    - If extra_cheese is True add 2 to bill.
    - If pepperoni is True add 3 to bill.
    - If olives is True add 1 to bill.
- **Output:** Print the final total bill.

```python
print('The Pizza Topping Calculator')

total_bill = 10.00

print(f'The cost of pizza is $ {total_bill}.')
print('Please select your toppings')

extra_cheese = input('Do you want extra cheese? (True/False): ').strip().capitalize() == 'True'
pepperoni = input('Do you want pepperoni? (True/False): ').strip().capitalize() == 'True'
olives = input('Do you want olives? (True/False): ').strip().capitalize() == 'True'

if extra_cheese:
    total_bill += 2.00
    print('Extra Cheese Topping: $', 2.00, sep='')

if pepperoni:
    total_bill += 3.00
    print('Pepperoni Topping: $', 3.00, sep='')

if olives:
    total_bill += 1.00
    print('Olives Topping: $', 1.00, sep='')

print(f'Your total bill amount is ${total_bill}')
```

**5. The Traffic Light Sensor** A self-driving car needs to interpret a single traffic light color.
- **Input:** User enters "Red", "Yellow", or "Green".
- **Scenario:**
    - If "Red" "Full Stop."
    - If "Yellow" "Slow down and prepare to stop."
    - If "Green" "Proceed with caution."
    - Anything else "Error: Signal Malfunction!"

```python
print('The Traffic Light Sensor')

traffic_light_colour = input('What is the present traffic light colour? ').strip().capitalize()

if (traffic_light_colour == 'Red'):
    print('Full Stop.')
elif (traffic_light_colour == 'Yellow'):
    print('Slow down and prepare to stop.')
elif (traffic_light_colour == 'Green'):
    print('Proceed with caution.')
else:
    print('Error: Signal Malfunction!')
```

**6. Movie Theatre Ticket Pricing** A theatre charges based on the age of the customer.
- **Input:** User enters their age.
- **Scenario:**
    - Age < 5 "Ticket is Free."
    - Age 5 to 17 "Child Ticket: $10."
    - Age 18 to 60 "Adult Ticket: $18."
    - Age > 60 "Senior Ticket: $12."

```python
print('Movie Theatre Ticket Pricing')

user_age = int(input('Enter your age: ').strip())

if (user_age < 5):
    print('Ticket is Free.')
elif (5 <= user_age <= 17):
    print('Child Ticket: $10.')
elif (18 <= user_age <= 60):
    print('Adult Ticket: $18.')
else:
    print('Senior Ticket: $12.')
```

**7. The 7-Day Steps Tracker** A fitness app wants to show your activity for the week.
- **Data:** A list of steps: *[4000, 7500, 10000, 2000, 8000, 12000, 5000]*.
- **Scenario:** Loop through the list and for each day, print: "Day [X]: You walked [Steps] steps."

```python
print('The 7-Day Steps Tracker')

steps = [4000, 7500, 10000, 2000, 8000, 12000, 5000]
day_counter = 1

for step_count in steps:
    print(f'Day {day_counter}: You walked {step_count} steps.')
    day_counter += 1
```
