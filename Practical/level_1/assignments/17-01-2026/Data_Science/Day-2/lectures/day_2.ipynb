## Type-Casting

### int to float, str, bool

#### Case 1:

When an integer is non-zero and positive the following happens:
- gets converted into float simply by passing the value in the float() function
- gets converted into string simply by passing the value in the str() function
- gets converted into boolean simply by passing the value in the bool() function and the value will be True

```python
number = 25
```

```python
number_float = float(number)
number_str = str(number)
number_bool = bool(number)
```

```python
print(f'The data type of {number} is {type(number)}')
print(f'The data type of {number_float} is {type(number_float)}')
print(f'The data type of {number_str} is {type(number_str)}')
print(f'The data type of {number_bool} is {type(number_bool)}')
```

#### Case 2:

When an integer is zero the following happens:
- gets converted into float simply by passing the value in the float() function
- gets converted into string simply by passing the value in the str() function
- gets converted into boolean simply by passing the value in the bool() function and the value will be False

```python
number_1 = 0
```

```python
number_1_float = float(number_1)
number_1_str = str(number_1)
number_1_bool = bool(number_1)
```

```python
print(f'The data type of {number_1} is {type(number_1)}')
print(f'The data type of {number_1_float} is {type(number_1_float)}')
print(f'The data type of {number_1_str} is {type(number_1_str)}')
print(f'The data type of {number_1_bool} is {type(number_1_bool)}')
```

#### Case 3:

When an integer is non-zero and negative the following happens:
- gets converted into float simply by passing the value in the float() function
- gets converted into string simply by passing the value in the str() function
- gets converted into boolean simply by passing the value in the bool() function and the value will be True

```python
negative_number = -25
```

### str to int, float, bool

#### Case 1:

When a string contains only numbers the following happens:
- gets converted into integer simply by passing the value in the int() function and the value is a positive integer
- gets converted into float simply by passing the value in the str() function and the value is a positive float
- gets converted into boolean simply by passing the value in the bool() function and the value will be True

```python
num = '2023'
```

```python
num_int = int(num)
num_float = float(num)
num_bool = bool(num)
```

```python
print(f'The data type of {num} is {type(num)}')
print(f'The data type of {num_int} is {type(num_int)}')
print(f'The data type of {num_float} is {type(num_float)}')
print(f'The data type of {num_bool} is {type(num_bool)}')
```

#### Case 2:

When a string contains numbers and only (.) as character the following happens:
- the int() function is not able to understand the character (.) and displays a Value Error on execution
- gets converted into float simply by passing the value in the str() function and the value is a positive float
- gets converted into boolean simply by passing the value in the bool() function and the value will be True

```python
num_1 = '2023.5'
```

```python
num_1_int = int(num_1)
```

```python
num_1_float = float(num_1)
num_1_bool = bool(num_1)
```

```python
print(f'The data type of {num_1} is {type(num_1)}')
print(f'The data type of {num_1_float} is {type(num_1_float)}')
print(f'The data type of {num_1_bool} is {type(num_1_bool)}')
```

#### Case 3:

When a string is empty the following happens:
- the int() function is not able to understand the blank string and displays a Value Error on execution
- the float() function is not able to understand the blank string and displays a Value Error on execution
- gets converted into boolean simply by passing the value in the bool() function and the value will be False

```python
string = ''
```

```python
string_int = int(string)
```

```python
string_float = float(string)
```

```python
string_bool = bool(string)
```

```python
print(f'The data type of {string} is {type(string)}')
print(f'The data type of {string_bool} is {type(string_bool)}')
```
