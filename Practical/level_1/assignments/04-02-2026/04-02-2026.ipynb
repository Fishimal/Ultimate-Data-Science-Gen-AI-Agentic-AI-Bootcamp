### Create an ultimate data science batch class and it should have the following properties:
    
        - Each batch will have its own instance
        - One batch should have only one instance
        - We should be able to add multiple students in a single batch along with their details
        - We should be able to print all the student names of that batch

```python
class DataScienceBatch:
    def __init__(self, batch_number, total_number_of_students=5):
        self.batch_number = batch_number
        self.total_number_of_students = total_number_of_students

        print(f'Welcome to {batch_number} Ultimate Data Science Bootcamp!')
    

    def input_student_details(self):
        self.student_details = {}

        for student_id in range(1, (self.total_number_of_students + 1)):
            name = ' '.join(input('Enter your name: ').split()).title()
            age = int(input('Enter your age: '))
            contact_number = input('Enter your phone number: ').strip()
            email_address = input('Enter your mail address: ').strip().lower()
            self.student_details[student_id] = {
                'Name': name,
                'Age': age,
                'Contact Number': contact_number,
                'Email': email_address
            }

            print(f'Thanks {name}, for registering in UDS {self.batch_number} Bootcamp.\nHappy Learning!\n')
    

    def get_all_student_names(self):
        print(f'\nFollowing are the student names enrolled in this {self.batch_number} UDS Bootcamp:')

        for student in self.student_details.keys():
            print(self.student_details[student]['Name'])
```

```python
print('Welcome to Krish Naik Academy')

batch_number = input('Enter the batch of Ultimate Data Science Bootcamp in which you want to register: ').strip()
```

```python
batch_1 = DataScienceBatch(batch_number)
```

```python
batch_1.input_student_details()
```

```python
batch_1.get_all_student_names()
```
