### Adding Content in Between a File

```python
import os
```

```python
file_name = 'temp.txt'
folder_name = 'temp'
file_path = os.path.join(os.getcwd(), folder_name, file_name)
```

- Creating a file using w+ mode such that if the file doesn't exist it gets created automatically.
- Writing contents in the file.
- Reading the contents of the file.

```python
with open(file_path, 'w+') as f:
    start_index = f.tell()
    print(f'The pointer is at location {start_index} in the file.')

    print('File is being written.')

    f.write(
        'Lorem ipsum dolor sit amet consectetur adipiscing elit. Quisque \nfaucibus ex sapien vitae pellentesque sem placerat. In id cursus mi \npretium tellus duis convallis. Tempus leo eu aenean sed diam urna \ntempor. Pulvinar vivamus fringilla lacus nec metus bibendum egestas. \nIaculis massa nisl malesuada lacinia integer nunc posuere. Ut hendrerit \nsemper vel class aptent taciti sociosqu. Ad litora torquent per conubia \nnostra inceptos himenaeos.\n\nLorem ipsum dolor sit amet consectetur adipiscing elit. Quisque \nfaucibus ex sapien vitae pellentesque sem placerat. In id cursus mi \npretium tellus duis convallis. Tempus leo eu aenean sed diam urna \ntempor. Pulvinar vivamus fringilla lacus nec metus bibendum egestas. \nIaculis massa nisl malesuada lacinia integer nunc posuere. Ut hendrerit \nsemper vel class aptent taciti sociosqu. Ad litora torquent per conubia \nnostra inceptos himenaeos.'
        )
    
    print('File writing is completed.')

    end_index = f.tell()
    print(f'The pointer is at location {end_index} in the file.\n')

    f.seek(start_index)
    get_all_contents = f.read()
    print(get_all_contents)
```

- Reading the created file using r+ mode.
- Fetching the index from where the new data needs to be added.
- Storing the end content of the file.
- Writing the new content to the file from the fetched index and also adding the end content of the file.

```python
with open(file_path, 'r+') as f:
    start_index = f.tell()
    get_all_contents = f.read()
    end_index = f.tell()
    line = 'nostra inceptos himenaeos.\n\n'
    line_start_index = get_all_contents.find(line)
    line_length = len(line)

    sub_content = get_all_contents[(line_start_index + line_length):]

    f.seek(line_start_index + line_length)
    f.write('New content added file modified\n\n' + sub_content)
```

- Reading the modified file in r mode.

```python
with open(file_path, 'r') as f:
    start_index = f.tell()
    print(f'The pointer is at location {start_index} in the file.\n')

    print('File is being read.\n')

    contents = f.read()
    print(contents)
    
    print('File reading is completed.\n')

    end_index = f.tell()
    print(f'The pointer is at location {end_index} in the file.\n')
```
