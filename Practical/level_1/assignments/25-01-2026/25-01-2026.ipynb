**1. The Smart Attendance Logger**

**The Goal:** Manage a classroom attendance list and ensure no duplicate entries are recorded if a student taps their ID card twice.
- **Concepts:** set , function , while.
- **The Problem:** Create a function mark_attendance() that keeps asking for student names until the user
types "STOP". Store names in a **Set** to automatically prevent duplicates.
- **Input:** Multiple strings (names) entered one by one.
- **Output:** The final count of unique students and the set itself.
- **Hint:**
    - **Expected Flow:** *Start an empty set Use a while True loop to take input() if name is "STOP", break
else add name to set Finally, print len(attendance_set).*

```python
def mark_attendance():
    attendance_set = set()
    name = ''

    while True:
        name = input('Enter your name (enter stop to exit): ').strip().upper()

        if (name == 'STOP'):
            break
        else:
            attendance_set.add(name)
    
    print('The below students are present:')

    for student in attendance_set:
        print(student)

    return len(attendance_set)


print('The Smart Attendance Logger')
total_number_of_unique_students = mark_attendance()
print(f'The total number of unique students are: {total_number_of_unique_students}')
```

**2. The Restaurant Bill Splitter**
- **The Goal:** Calculate how much each person owes, including a tax and a custom tip percentage.
- **Concepts:** functions, default values, list.
- **The Problem:** Write a function split_bill(items_prices, people, tax=0.05, tip=0.10). items_prices will be a List of decimals (floats).
- **Input:** A list like [25.0, 15.0, 10.0] and an integer for people.
- **Output:** A single float (the amount per person).
- **Hint:**
    - **Expected Flow:** *Calculate sum() of the list Add tax and tip amounts Divide the grand total by people return the result.*

```python
def split_bill(item_prices, people, tax=0.05, tip=0.10):
    sum_of_item_prices = sum(item_prices)
    taxed_amount = (sum_of_item_prices * tax)
    tip_amount = (sum_of_item_prices * tip)
    total_amount = (sum_of_item_prices + taxed_amount + tip_amount)
    individual_amount = (total_amount / people)
    print(f'Sum of price for all items: ${sum_of_item_prices}')
    print(f'Tax ({tax*100}%): ${taxed_amount:.2f}')
    print(f'Tip ({tip*100}%): ${tip_amount:.2f}')
    print(f'Final bill amount including tax: ${total_amount}')
    return individual_amount


print('The Restaurant Bill Splitter')
item_prices = []
people = int(input('Enter the total number of people: ').strip())

for i in range(3):
    item_price = float(input(f'Enter the cost of item_{(i+1)}: ').strip())
    item_prices.append(item_price)
    print(f'Price of item_{i+1}: ${item_price}')

individual_bill_amount = split_bill(item_prices, people)

print(f'The bill amount is split into {people} people. Each one needs to pay ${individual_bill_amount:.2f}')
```

**4. The E-Commerce Discount Engine**
- **The Goal:** Apply different discount rates based on the total purchase amount.
- **Concepts:** if-elif-else , functions.
- **The Problem:** Create a function apply_discount(total).
    - If total > $500: 20% off.
    - If total > $200: 10% off.
    - Else: No discount.
- **Input:** A single number (the total bill).
- **Output:** The final price after the discount is subtracted.
- **Hint:**
    - **Expected Flow:** *Receive the total Use an if-elif-else chain to determine the discount percentage Calculate the new price return it.*

```python
def apply_discount(total):
    net_price = None
    discount_percentage = 0

    if (total > 500):
        discount_percentage = 20
        print(f'Since your total bill amount is greater than $500.00, you get a discount of {discount_percentage}%.')
        net_price = (total * (1 - (discount_percentage / 100)))
    elif (total > 200):
        discount_percentage = 10
        print(f'Since your total bill amount is greater than $200.00, you get a discount of {discount_percentage}%.')
        net_price = (total * (1 - (discount_percentage / 100)))
    else:
        print(f'Since your total bill amount is less than or equal to $200.00, you are not eligible for a discount.')
    
    return net_price


print('The E-Commerce Discount Engine')

total_bill_amount = float(input('Enter your total bill amount: ').strip())

print(f'Your total bill amount is: ${total_bill_amount}')

discounted_price = apply_discount(total_bill_amount)

if discounted_price is not None:
    print(f'Your bill amount after applying discount: ${discounted_price:.2f}')
else:
    print(f'Your final bill amount: ${total_bill_amount:.2f}')
```

**7. The ATM Pin Validator**
- **The Goal:** Lock the user out after 3 incorrect attempts.
- **Concepts:** while, if-else, break.
- **The Problem:** Set correct_pin = "1234". Allow the user 3 tries to enter the right PIN.
- **Input:** Up to 3 string inputs.
- **Output:** "Welcome" on success, or "Card Blocked" after the 3rd fail.
- **Hint:**
    - **Expected Flow:** *Set attempts = 0 While attempts < 3 : Take PIN input if correct, print "Welcome" and break else increment attempts . If attempts reaches 3, print "Card Blocked."*

```python
print('The ATM Pin Validator')

atm_pin = ''
attempts = 0
correct_pin = '1234'

while True:
    if attempts < 3:
        atm_pin = input('Enter your ATM pin: ').strip()
        if (atm_pin == correct_pin):
            print('Welcome')
            break
        else:
            attempts += 1
    else:
        print('Card Blocked')
        break
```

**9. The Grocery Inventory Update**
- **The Goal:** Update the quantity of items in a store after a shipment arrives.
- **Concepts:** dict, for loop.
- **The Problem:** Given inventory = {"Apples": 50, "Bananas": 20} , a shipment of 10 more of each arrives. Update the dictionary using a loop.
- **Input:** None.
- **Output:** The updated dictionary.
- **Hint:**
    - **Expected Flow:** *Loop through inventory.keys() For each item, add 10 to the current value Print the dictionary after the loop finishes.*


```python
inventory = {'Apples': 50, 'Bananas': 20}

print('The Grocery Inventory Update')
print(f'Stocks before updation: {inventory}')

for fruit in inventory.keys():
    inventory[fruit] += 10

print(f'Stocks after updation: {inventory}')
```

**10. The Flexible Event Invitation**
- **The Goal:** Create a function that can accept any number of guests and specific event details.
- **Concepts:** functions, *args, **kwargs.
- **The Problem:** Write create_invite(event_name, *guest_names, **details).
- **Input:** One string, followed by multiple guest names, followed by named details (e.g. Time="6pm" ).
- **Output:** A printed invitation showing the event, guests, and details.
- **Hint:**
    - *Expected Flow: Print the event_name Use a for loop to iterate through \*args (guests) Use .items() to loop through \*\*details (extra info) and print them.*

```python
def create_invite(event_name, *guest_names, **details):
    print(f'Invitation for {event_name}')
    print('Below is the guest list:')

    for index, guest in enumerate(guest_names, start=1):
        print(f'{index}. {guest}')

    for key, value in details.items():
        print(f'{key} : {value}')


print('The Flexible Event Invitation')

event_name = input('Enter the event name: ').strip().title()
number_of_guests = int(input('Enter the number of guests: ').strip())
guest_list = []
event_details = {}

for i in range(number_of_guests):
    guest_name = input('Enter the guest name: ').strip().title()
    guest_list.append(guest_name)

while True:
    key = input('Enter key (or quit to stop): ').strip().title()

    if (key.lower() == 'quit'):
        break
    elif key in event_details:
        continue
    else:
        value = input(f'Enter the {key}: ').strip().title()
        event_details[key] = value

create_invite(event_name, *guest_list, **event_details)

```
